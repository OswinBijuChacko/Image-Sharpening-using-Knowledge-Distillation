```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms
import cv2
import numpy as np
import os
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim
import lpips
import urllib.request
import zipfile

# Setup
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
os.makedirs('data', exist_ok=True)

# 1. Dataset Preparation
def download_dataset():
    if not os.path.exists('data/DIV2K_train_HR'):
        print("Downloading dataset...")
        url = "https://data.vision.ee.ethz.ch/cvl/DIV2K/DIV2K_train_HR.zip"
        urllib.request.urlretrieve(url, 'data/DIV2K_train_HR.zip')
        with zipfile.ZipFile('data/DIV2K_train_HR.zip', 'r') as zip_ref:
            zip_ref.extractall('data/')
        os.remove('data/DIV2K_train_HR.zip')

download_dataset()

class SharpeningDataset(Dataset):
    def __init__(self, root_dir, patch_size=128, train=True):
        self.image_files = [os.path.join(root_dir, f) for f in os.listdir(root_dir)
                          if f.lower().endswith(('.png', '.jpg', '.jpeg'))][:800]
        self.patch_size = max(patch_size, 64)
        self.train = train
        self.to_tensor = transforms.ToTensor()

    def __len__(self): return len(self.image_files)

    def __getitem__(self, idx):
        img = np.array(Image.open(self.image_files[idx]).convert('RGB'))
        h, w = img.shape[:2]

        if self.train and h > self.patch_size and w > self.patch_size:
            x = np.random.randint(0, w - self.patch_size)
            y = np.random.randint(0, h - self.patch_size)
            img = img[y:y+self.patch_size, x:x+self.patch_size]
        else:
            img = cv2.resize(img, (self.patch_size, self.patch_size))

        # Random blur strength during training
        sigma = np.random.uniform(0.8, 1.5) if self.train else 1.5
        blurred = cv2.GaussianBlur(img, (5, 5), sigmaX=sigma)
        return self.to_tensor(blurred).float(), self.to_tensor(img).float()

# 2. Enhanced Model Architectures
class TeacherNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.initial = nn.Sequential(
            nn.Conv2d(3, 128, 9, padding=4),
            nn.PReLU()
        )
        self.res_blocks = nn.Sequential(
            *[nn.Sequential(
                nn.Conv2d(128, 128, 3, padding=1),
                nn.BatchNorm2d(128),
                nn.PReLU(),
                nn.Conv2d(128, 128, 3, padding=1),
                nn.BatchNorm2d(128)
            ) for _ in range(8)]
        )
        self.final = nn.Conv2d(128, 3, 9, padding=4)

    def forward(self, x):
        x = self.initial(x)
        res = x
        x = self.res_blocks(x)
        return torch.sigmoid(self.final(x + res))

class StudentNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.initial = nn.Sequential(
            nn.Conv2d(3, 64, 7, padding=3),
            nn.PReLU()
        )
        self.res_blocks = nn.Sequential(
            *[nn.Sequential(
                nn.Conv2d(64, 64, 3, padding=1),
                nn.BatchNorm2d(64),
                nn.PReLU(),
                nn.Conv2d(64, 64, 3, padding=1),
                nn.BatchNorm2d(64)
            ) for _ in range(4)]
        )
        self.final = nn.Conv2d(64, 3, 7, padding=3)

    def forward(self, x):
        x = self.initial(x)
        res = x
        x = self.res_blocks(x)
        return torch.sigmoid(self.final(x + res))

# 3. Metric Calculation with Proper Detachment
def calculate_metrics(output, target, lpips_fn):
    metrics = {'psnr': 0, 'ssim': 0, 'lpips': 0}
    batch_size = output.shape[0]

    output_detached = output.detach().cpu()
    target_detached = target.detach().cpu()

    for i in range(batch_size):
        # Convert to numpy (HWC format)
        output_np = output_detached[i].permute(1, 2, 0).numpy()
        target_np = target_detached[i].permute(1, 2, 0).numpy()

        # Calculate metrics
        metrics['psnr'] += psnr(target_np, output_np, data_range=1.0)
        win_size = min(7, min(output_np.shape[0], output_np.shape[1]) - 1)
        win_size = win_size if win_size % 2 == 1 else win_size - 1
        metrics['ssim'] += ssim(target_np, output_np,
                              win_size=win_size,
                              channel_axis=2,
                              data_range=1.0)
        metrics['lpips'] += lpips_fn(
            output[i].unsqueeze(0)*2-1,
            target[i].unsqueeze(0)*2-1
        ).item()

    # Average across batch
    for k in metrics:
        metrics[k] /= batch_size

    # Composite accuracy (87-90% target)
    metrics['accuracy'] = 100 * (0.4*(metrics['psnr']/40) + 0.4*metrics['ssim'] + 0.2*(1-metrics['lpips']))
    return metrics

# 4. Teacher Training
def train_teacher(teacher, train_loader, val_loader, epochs=30):
    teacher.train()
    optimizer = optim.AdamW(teacher.parameters(), lr=1e-4, weight_decay=1e-4)
    criterion = nn.L1Loss()
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=3, factor=0.5)
    lpips_fn = lpips.LPIPS(net='alex').to(device)

    best_accuracy = 0
    for epoch in range(epochs):
        teacher.train()
        train_metrics = {'psnr': 0, 'ssim': 0, 'lpips': 0, 'accuracy': 0}

        for blurred, sharp in tqdm(train_loader, desc=f"Teacher Epoch {epoch+1}"):
            blurred, sharp = blurred.to(device), sharp.to(device)

            optimizer.zero_grad()
            output = teacher(blurred)
            loss = criterion(output, sharp)
            loss.backward()
            optimizer.step()

            metrics = calculate_metrics(output, sharp, lpips_fn)
            for k in train_metrics:
                train_metrics[k] += metrics[k]

        # Validation
        teacher.eval()
        val_metrics = {'psnr': 0, 'ssim': 0, 'lpips': 0, 'accuracy': 0}
        with torch.no_grad():
            for blurred, sharp in val_loader:
                blurred, sharp = blurred.to(device), sharp.to(device)
                output = teacher(blurred)
                metrics = calculate_metrics(output, sharp, lpips_fn)
                for k in val_metrics:
                    val_metrics[k] += metrics[k]

        # Average metrics
        for k in train_metrics:
            train_metrics[k] /= len(train_loader)
            val_metrics[k] /= len(val_loader)

        scheduler.step(val_metrics['psnr'])

        print(f"\nTeacher Epoch {epoch+1} Results:")
        print(f"Train - PSNR: {train_metrics['psnr']:.2f} dB | SSIM: {train_metrics['ssim']:.4f} | LPIPS: {train_metrics['lpips']:.4f} | Acc: {train_metrics['accuracy']:.2f}%")
        print(f"Val   - PSNR: {val_metrics['psnr']:.2f} dB | SSIM: {val_metrics['ssim']:.4f} | LPIPS: {val_metrics['lpips']:.4f} | Acc: {val_metrics['accuracy']:.2f}%")

        if val_metrics['accuracy'] > best_accuracy:
            best_accuracy = val_metrics['accuracy']
            torch.save(teacher.state_dict(), 'teacher.pth')
            print(f"New best teacher saved! (Accuracy: {best_accuracy:.2f}%)")

    print(f"Teacher training complete. Best Accuracy: {best_accuracy:.2f}%")

# 5. Knowledge Distillation Training
def train_kd(teacher, student, train_loader, val_loader, epochs=100):
    teacher.eval()
    student.train()

    optimizer = optim.AdamW(student.parameters(), lr=1e-4, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs)

    criterion_mse = nn.MSELoss()
    criterion_l1 = nn.L1Loss()
    lpips_fn = lpips.LPIPS(net='alex').to(device)

    best_accuracy = 0
    for epoch in range(epochs):
        student.train()
        train_metrics = {'psnr': 0, 'ssim': 0, 'lpips': 0, 'accuracy': 0}

        for blurred, sharp in tqdm(train_loader, desc=f"KD Epoch {epoch+1}"):
            blurred, sharp = blurred.to(device), sharp.to(device)

            with torch.no_grad():
                teacher_out = teacher(blurred)

            optimizer.zero_grad()
            student_out = student(blurred)

            # Combined loss
            loss = 0.7 * criterion_mse(student_out, teacher_out) + 0.3 * criterion_l1(student_out, sharp)
            loss.backward()
            optimizer.step()

            metrics = calculate_metrics(student_out, sharp, lpips_fn)
            for k in train_metrics:
                train_metrics[k] += metrics[k]

        scheduler.step()

        # Validation
        student.eval()
        val_metrics = {'psnr': 0, 'ssim': 0, 'lpips': 0, 'accuracy': 0}
        with torch.no_grad():
            for blurred, sharp in val_loader:
                blurred, sharp = blurred.to(device), sharp.to(device)
                output = student(blurred)
                metrics = calculate_metrics(output, sharp, lpips_fn)
                for k in val_metrics:
                    val_metrics[k] += metrics[k]

        # Average metrics
        for k in train_metrics:
            train_metrics[k] /= len(train_loader)
            val_metrics[k] /= len(val_loader)

        print(f"\nKD Epoch {epoch+1} Results:")
        print(f"Train - PSNR: {train_metrics['psnr']:.2f} dB | SSIM: {train_metrics['ssim']:.4f} | LPIPS: {train_metrics['lpips']:.4f} | Acc: {train_metrics['accuracy']:.2f}%")
        print(f"Val   - PSNR: {val_metrics['psnr']:.2f} dB | SSIM: {val_metrics['ssim']:.4f} | LPIPS: {val_metrics['lpips']:.4f} | Acc: {val_metrics['accuracy']:.2f}%")

        if val_metrics['accuracy'] > best_accuracy:
            best_accuracy = val_metrics['accuracy']
            torch.save(student.state_dict(), 'best_student.pth')
            print(f"New best student saved! (Accuracy: {best_accuracy:.2f}%)")

            # Visual comparison
            test_blurred, test_sharp = next(iter(val_loader))
            test_blurred = test_blurred[:1].to(device)
            with torch.no_grad():
                teacher_out = teacher(test_blurred).cpu()
                student_out = student(test_blurred).cpu()

            plt.figure(figsize=(18,5))
            titles = ['Blurred', 'Teacher', 'Student', 'Sharp']
            images = [
                test_blurred[0].cpu(),
                teacher_out[0],
                student_out[0],
                test_sharp[0]
            ]
            for i, (title, img) in enumerate(zip(titles, images)):
                plt.subplot(1,4,i+1)
                plt.imshow(img.permute(1,2,0).clip(0,1))
                plt.title(title)
                plt.axis('off')
            plt.tight_layout()
            plt.show()

    print(f"KD training complete. Best Accuracy: {best_accuracy:.2f}%")

# 6. Main Execution
if __name__ == "__main__":
    # Prepare data
    full_dataset = SharpeningDataset('data/DIV2K_train_HR', train=True)
    train_size = int(0.9 * len(full_dataset))
    val_size = len(full_dataset) - train_size
    train_dataset, val_dataset = torch.utils.data.random_split(full_dataset, [train_size, val_size])

    train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, num_workers=4, pin_memory=True)
    val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False, num_workers=4, pin_memory=True)

    # Initialize models
    teacher = TeacherNet().to(device)
    student = StudentNet().to(device)

    # Train or load teacher
    if os.path.exists('teacher.pth'):
        teacher.load_state_dict(torch.load('teacher.pth'))
        print("Loaded pretrained teacher model")
    else:
        print("Training teacher model...")
        train_teacher(teacher, train_loader, val_loader, epochs=30)

    # Knowledge distillation
    print("\nStarting knowledge distillation...")
    train_kd(teacher, student, train_loader, val_loader, epochs=100)

    # Final test
```

```python
print("\nRunning final test...")
urllib.request.urlretrieve(
    "https://upload.wikimedia.org/wikipedia/commons/thumb/5/50/Vd-Orig.png/800px-Vd-Orig.png",
    "test_image.png"
)

# Load model
student.load_state_dict(torch.load('best_student.pth', map_location=device))
student = student.to(device)
student.eval()

# Prepare images
img = Image.open("test_image.png").convert('RGB')
img_tensor = transforms.ToTensor()(img).unsqueeze(0).to(device)
blurred = cv2.GaussianBlur(np.array(img), (5,5), sigmaX=1.5)
blurred_tensor = transforms.ToTensor()(blurred).unsqueeze(0).to(device)

# Process image
with torch.no_grad():
    sharpened = student(blurred_tensor).clamp(0, 1)

# Convert to numpy for SSIM calculation
def tensor_to_np(tensor):
    return tensor.squeeze(0).permute(1, 2, 0).cpu().numpy()

sharpened_np = tensor_to_np(sharpened)
target_np = tensor_to_np(img_tensor)

# Calculate metrics
def safe_psnr(output, target):
    mse = torch.mean((output - target)**2)
    return -10 * torch.log10(mse + 1e-8)

psnr_value = min(50.0, safe_psnr(sharpened, img_tensor).item())
ssim_value = min(1.0, ssim(target_np, sharpened_np,
                         win_size=7,
                         channel_axis=2,
                         data_range=1.0))

lpips_fn = lpips.LPIPS(net='alex').to(device)
lpips_value = min(1.0, max(0.0, lpips_fn(sharpened*2-1, img_tensor*2-1).item()))

# Calculate bounded accuracy
accuracy = 100 * (
    0.4 * (psnr_value / 50) +  # Normalized PSNR
    0.4 * ssim_value +        # SSIM
    0.2 * (1 - lpips_value)   # Inverted LPIPS
)
accuracy = min(100.0, max(0.0, accuracy))

# Visualization
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.imshow(blurred)
plt.title("Blurred Input")
plt.subplot(1,3,2)
plt.imshow(sharpened_np.clip(0,1))
plt.title(f"Sharpened\nPSNR: {psnr_value:.2f} dB | SSIM: {ssim_value:.4f}")
plt.subplot(1,3,3)
plt.imshow(img)
plt.title("Original")
plt.show()

print(f"\nFinal Test Metrics:")
print(f"PSNR: {psnr_value:.2f} dB (Higher better)")
print(f"SSIM: {ssim_value:.4f} (Higher better)")
print(f"LPIPS: {lpips_value:.4f} (Lower better)")
print(f"Accuracy: {accuracy:.2f}% (Properly bounded)")
```
